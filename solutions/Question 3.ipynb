## 2. Calculate MACD (slow/medium/fast) and RSI for each metal historically

I have summarised each function, and its associated parameters, as well as any corresponding mathematical interpretations before defining each function.

### 2.1 Calculating the exponential moving average (EMA)

- The cell below defines the function **`calculate_ema()`** which returns the exponential moving average (EMA) required for calculations of the MACD. 
- **`calculate_ema()`** has two parameters; the first parameter **`prices`** expects a DataFrame while the second parameter **`span`** expects an integer.
- The parameter **`span`** refers to the number of data points assigned significant weighting within the EMA, i.e. a 12-day EMA assigns significant weighting to daily price data for the past 12 days and the associated weighting for data points before this window decreases exponentially.
- Calculating the MACD requires evaluating various EMAs. Thus **`calculate_ema()`** taking the variable parameter **`span`** promotes code  reusability.
- The built-in Pandas function **`emw()`**, which provides exponentially weighted calculations, is called in the function body of **`calculate_ema()`**. 
- The **`span`** parameter associated to **`emw()`** specifies the decay rate via the formula for the smoothing parameter $ \alpha $, namely: $\alpha = \frac{2}{\textrm{span}+1} $. It is passed into **`emw()`**, via **`span=span`**, from the integer value passed into **`calculate_ema()`** when the latter function is called.
- In financial time-series analysis, EMAs are usually calculated without normalising the weights which is achieved by setting **`adjust=False`**.
- The final EMA is calculated, from the exponentially weighted moving object returned by **`emw()`**, by averaging using the **`mean()`** function.


```python
# Returns the n-day EMA where n = span 
def calculate_ema(prices, span):
    return prices.ewm(span=span, adjust=False).mean()
```

### 2.2 Calculating the moving average convergence/divergence (MACD)

- The cell below defines the function **`calculate_macd()`** which returns the moving average convergence/divergence (MACD) line and the signal line. 
- **`calculate_macd()`** has four parameters; the first parameter **`prices`** expects a DataFrame while the remaining parameters, i.e. **`fast_period`**, **`slow_period`** and **`signal_period`**, expect integers. The fast-period EMA and slow-period EMA are the fast line and the slow line respectively.
- The primary line of the MACD, known as the MACD line, is calculated by subtracting the slow line from the fast line.
- The MACD line also gives rise to the signal line, which is an EMA of the MACD line itself over the **`signal_period`**.
- The function checks that the returned values are numeric pandas Series since the **`round()`** method will subsequently be applied to them.

```python
# Returns the MACD and the signal line 
def calculate_macd(prices, fast_period, slow_period, signal_period):
    
    # Calculates the fast-line
    ema_fast = calculate_ema(prices, fast_period)
    # Calculates the slow-line
    ema_slow = calculate_ema(prices, slow_period)
    
    # Calculates the MACD (difference between fast-line and slow-line)
    macd = ema_fast - ema_slow
    
    # Calculates the signal line
    signal = calculate_ema(macd, signal_period)
    
    # Checks to ensure returned values are numeric pandas Series
    assert isinstance(macd, pd.Series), "MACD is not a pandas Series"
    assert isinstance(signal, pd.Series), "Signal is not a pandas Series"
    assert pd.api.types.is_numeric_dtype(macd), "MACD Series is not numeric"
    assert pd.api.types.is_numeric_dtype(signal), "Signal Series is not numeric"
    
    return macd, signal
```

#### 2.2.1 Determining MACD (slow/ medium/ fast) settings
- Need to determine which integer values to use in calculating the fast-line, the slow-line and the signal-line in each of the slow/ medium/ fast MACDs.
- The typical MACD settings has the following values: fast-line - 12,  slow-line - 26 and signal-line - 9. These settings can be used for the medium MACD.
- The fast MACD reacts more quickly to price changes and is useful for identifying short-term movements: fast-line - 5,  slow-line - 13 and signal-line - 9.
- The slow MACD is more stable; it is often used to identify longer-term trends and to filter out noise: fast-line - 19,  slow-line - 39 and signal-line - 9.

### 2.3 Calculating the relative strength index (RSI)

- The cell below defines the function **`calculate_rsi()`** which takes two parameters and returns the relative strength index (RSI).
- The first parameter **`prices`** expects a DataFrame while the second parameter **`periods`** expects an integer (with default value of 14 if not specified).
- The function uses the **`diff()`** method on **`prices`** to calculate the difference between successive data points and stores it in the variable **`delta`**.
- The positive (gains) and negative (losses) price changes are separated from **`delta`** using the **`where()`** method, which replaces the price changes with zeros if they are negative and positive respectively, maintaining index structure. **`fillna(0)`** replaces any **`NaN`** values with zeros if any remain.
- The average gain/loss is calculated by using **`mean()`** to average the values within the **`window`** generated by the **`rolling`** method, the default length of which is set to 14 via **`windows=periods`**. **`min_periods`** ensures that the average is only calculated when there are enough data points.
- The relative strength is prevented from causing a **`ZeroDivisionError`** during a period of no losses by adding a small quantity to the average loss.
- The function checks that the returned RSI value is a numeric pandas Series since the **`round()`** method will subsequently be applied to it.

```python
# Returns the RSI 
def calculate_rsi(prices, periods=14):
    
    # Calculates the price difference between successive data points
    delta = prices.diff()
    
    # Creates a series of positive changes only
    gain = (delta.where(delta > 0, 0)).fillna(0)
    # Creates a series of negative changes only
    loss = (-delta.where(delta < 0, 0)).fillna(0)
    
    # Calculates the average gain over the window
    avg_gain = gain.rolling(window=periods, min_periods=periods).mean()
    # Calculates the average loss over the window
    avg_loss = loss.rolling(window=periods, min_periods=periods).mean()

    # Calculates the relative strength as (average gain/ average loss) 
    rs = avg_gain / (avg_loss + 1e-10)
    
    # Converts relative strength into RSI
    rsi = 100 - (100 / (1 + rs))
    
    # Checks to ensure returned values are numeric pandas Series
    assert isinstance(rsi, pd.Series), "RSI is not a pandas Series"
    assert pd.api.types.is_numeric_dtype(rsi), "RSI Series is not numeric"
    
    return rsi
```

### 2.4 Results

The results are generated by first creating a **`metals`** list as well as a **`results`** dictionary. Each element of the list **`metals`** (i.e. **`Copper`** and **`Zinc`**) along with a corresponding DataFrame will form a key-value pair in the **`results`** dictionary. Then, we iterate over each element in **`metals`** where:
- The price data in the **`filtered_data`** DataFrame above is assigned to the variable **`prices`**.
- The fast, medium and slow MACDs are calculated as well as their corresponding signal lines  together with the RSI.
- A DataFrame, itself consisting of key-value pairs, is created where the keys represent columns and the values are the data populating those columns.
- This DataFrame is then stored in the **`results`** dictionary as the value associated to a key that is the element of **`metals`** currently being processed.

The result is a **`results`** dictionary where each key is a metal and each value is a DataFrame corresponding to that metal. We then check that all required quantities are being calculated and stored correctly by printing the first and last records in the corresponding DataFrame for each metal in **`results`**.

```python
# Creates a list of metals to iterate over
metals = ['Copper', 'Zinc']

# Creates an empty dictionary to store the generated data
results = {}
```

```python
# Loops over the list of metals
for metal in metals:
    
    # Assigns price data in filtered_data for the current metal to the variable prices
    prices = filtered_data[metal]
    
    # Calculates fast MACD and signal line
    macd_fast, signal_fast = calculate_macd(prices, 5, 13, 9)
    # Calculates medium MACD and signal line
    macd_medium, signal_medium = calculate_macd(prices, 12, 26, 9)
    # Calculates slow MACD and signal line
    macd_slow, signal_slow = calculate_macd(prices, 19, 39, 9)
    
    # Calculates and stores RSI
    rsi = calculate_rsi(prices)
    
    # Creates a DataFrame for the current metal and stores it in the results dictionary with metal as the key
    results[metal] = pd.DataFrame({
        'Date': filtered_data['Date'],
        'Metal': metal,
        'Price': prices,
        'MACD_Fast': macd_fast.round(2),
        'Signal_Fast': signal_fast.round(2),
        'MACD_Medium': macd_medium.round(2),
        'Signal_Medium': signal_medium.round(2),
        'MACD_Slow': macd_slow.round(2),
        'Signal_Slow': signal_slow.round(2),
        'RSI': rsi.round(2)
    })
```

```python
# Prints first and last records in the corresponding DataFrame for each metal
for metal, df in results.items():
    print(f"Data for {metal}:\n")
    print(f"{df.head(4)}\n")
    print(f"{df.tail(4)}\n")
```

The data generated above needs to populate a SQL table, which generally expects records of data be inserted row by row. A convenient method is to flatten a DataFrame into a list of tuples especially since tuples are immutable objects, which ensure data integrity by preventing accidental alteration before insertion into the table. An empty list **`data_to_insert`** is created to store each tuple. Then, iterating over each (metal, DataFrame) pair in **`results`**:

- The DateTime object in the **`Date`** column of the DataFrame is first converted into the standard SQL date format (yyyy-mm-dd).
- The DataFrame (value) is flattened while simultaneously inserting the name of the corresponding metal (key) into the second element of each tuple.

We then check that the list of tuples is being correctly generated and each tuple is correctly formatted by printing the first few tuples in the list.

```python
formatted_results = {}

for metal, df in results.items():
    
    # Converts 'Date' column to string formatted for SQL
    df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
    
    # Stores the formatted DataFrame in the new dictionary
    formatted_results[metal] = df
```

## 4. Demonstrate the use of a decorator to log the execution of the SQL inserts

```python
# Logs the execution of SQL inserts
def log_sql_operation(func):
    @functools.wraps(func)
    def wrapper_decorator(*args, **kwargs):
        result = func(*args, **kwargs)
        print(f"Logged SQL operation: {func.__name__}, Date: {args[0]}")
        return result
    return wrapper_decorator
```

```python
# Applies the decorator to insert function
@log_sql_operation
def insert_data(date, metal, price, 
                macd_fast, signal_fast, 
                macd_medium, signal_medium, 
                macd_slow, signal_slow, 
                rsi):

    # Inserts values
    c.execute('''
    INSERT INTO metal_analysis (
        Date, Metal, Price, 
        MACD_Fast, Signal_Fast, 
        MACD_Medium, Signal_Medium, 
        MACD_Slow, Signal_Slow, RSI)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    ''', (date, metal, price, 
          macd_fast, signal_fast, 
          macd_medium, signal_medium, 
          macd_slow, signal_slow, 
          rsi))
    
    # Commits the changes
    conn.commit()
```

```python
# Inserts each record in list of tuples into metal_analysis
for record in data_to_insert:
    # Unpacks each tuple into variables
    (date, metal, price, 
     macd_fast, signal_fast, 
     macd_medium, signal_medium, 
     macd_slow, signal_slow, 
     rsi) = record

    # Converts NaN to None for SQL compatibility
    rsi = None if rsi != rsi else rsi

    # Calls the insert_data function with the unpacked variables
    insert_data(date, metal, price, 
                macd_fast, signal_fast, 
                macd_medium, signal_medium, 
                macd_slow, signal_slow, 
                rsi)
```
